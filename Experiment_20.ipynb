**DATA ANALYSIS**

```python
from sklearn.preprocessing import StandardScaler
# Tạo X và y từ DataFrame df
X = df.drop("Credit_Score",axis=1).values
y = df["Credit_Score"].values
# Khởi tạo scaler
scaler = StandardScaler()
# Scale các features trong X
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from imblearn.over_sampling import RandomOverSampler
# Khởi tạo RandomOverSampler
oversampling = RandomOverSampler(random_state=42)
# Cân bằng lớp cho X_scaled và y
X_resampled, y_resampled = oversampling.fit_resample(X_scaled, y)
# Hiển thị kích thước của X_resampled và y_resampled sau khi cân bằng lớp
print("Kích thước của X_resampled:", X_resampled.shape)
print("Kích thước của y_resampled:", y_resampled.shape)
```

# **XÂY DƯNG MÔ HÌNH**

# **Splitting the data**

```python
from sklearn.model_selection import StratifiedKFold
from keras.models import Sequential
from keras.layers import Dense, Dropout, BatchNormalization
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import numpy as np
import time
from keras.metrics import Precision, Recall, F1Score
```

```python
# Define number of folds for K-Fold cross-validation
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
```

```python
#Baseline MLP
# Initialize lists to store evaluation metrics for each fold
accuracy_scores = []
f1_scores = []
precision_scores = []
recall_scores = []
execution_times = []

# Iterate over each fold
for train_index, test_index in skf.split(X_resampled, y_resampled):
  # Split data into train and test sets for this fold
  X_train_fold, X_test_fold = X_resampled[train_index], X_resampled[test_index]
  y_train_fold, y_test_fold = y_resampled[train_index], y_resampled[test_index]

  # Initialize the model
  model = Sequential()
  # Add input layer
  model.add(Dense(64, activation='relu', input_shape=(X_resampled.shape[1],)))
  # Add hidden layers
  model.add(Dense(128, activation='relu'))
  model.add(Dropout(0.2)) # Adding dropout for regularization
  model.add(Dense(64, activation='relu'))

  # Add output layer
  model.add(Dense(3, activation='softmax')) # Assuming 3 classes for credit score

  # Compile the model
  model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

  # Train the model and measure execution time
  start_time = time.time()
  history = model.fit(X_train_fold, y_train_fold, epochs=20, batch_size=32, verbose=0)
  end_time = time.time()
  execution_time = end_time - start_time
  execution_times.append(execution_time)

  # Predict labels for the test data of current fold
  y_pred_proba = model.predict(X_test_fold)
  y_pred = np.argmax(y_pred_proba, axis=1)

  # Calculate evaluation metrics for the current fold
  accuracy = accuracy_score(y_test_fold, y_pred)
  f1 = f1_score(y_test_fold, y_pred, average='macro')
  precision = precision_score(y_test_fold, y_pred, average='macro')
  recall = recall_score(y_test_fold, y_pred, average='macro')

  # Append evaluation metrics to the respective lists
  accuracy_scores.append(accuracy)
  f1_scores.append(f1)
  precision_scores.append(precision)
  recall_scores.append(recall)

# Calculate and print average evaluation metrics and execution time across all folds
avg_accuracy = sum(accuracy_scores) / n_splits
avg_f1 = sum(f1_scores) / n_splits
avg_precision = sum(precision_scores) / n_splits
avg_recall = sum(recall_scores) / n_splits
avg_execution_time = sum(execution_times) / n_splits
print("Accuracy:", avg_accuracy)
print("Precision:", avg_precision)
print("Recall:", avg_recall)
print("F1-Score:", avg_f1)
print("Execution Time (seconds):", avg_execution_time)
```
